# Feature selection 

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical_scaled = scaler.fit_transform(numerical)
```

```python
pd.DataFrame(numerical_scaled).head()
```

```python
from sklearn.feature_selection import VarianceThreshold 

var_threshold = 0.02
sel = VarianceThreshold(threshold=(var_threshold))

sel = sel.fit(numerical_scaled)
temp = sel.transform(numerical_scaled)
temp = pd.DataFrame(temp)
print(numerical_scaled.shape)
print(temp.shape)
```

```python
pd.DataFrame(temp).head()
```

```python
#the variance of each feature in order of appearance of the dataset.

sel.variances_
```

```python
#The result as True and False for the columns that we selected 
sel.variances_ > var_threshold
sel.get_support()
var_list = list(sel.get_support())
var_list
```

```python
# combine var_list with the columns list using zip

list(zip(numerical.columns, var_list))
```

```python
#list of low_variance columns, to be dropped

[col[0] for col in zip(numerical.columns, var_list) if col[1] == False]
```

```python
drop_list= [col[0] for col in zip(numerical.columns, var_list) if col[1] == False]
```

```python
numerical = numerical.drop(drop_list, axis=1)
```

# Random Forests 

```python
df = df.drop(['TARGET_D'], axis=1)
df
```

```python
df['TARGET_B'].value_counts()
```

```python
df.isna().sum().sum()
```

```python
y = df['TARGET_B']
X = df.drop(['TARGET_B'], axis = 1)

numericalX = X.select_dtypes(np.number)
categoricalX = X.select_dtypes(np.object)


from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first').fit(categoricalX)
encoded_categorical = encoder.transform(categoricalX).toarray()
encoded_categorical = pd.DataFrame(encoded_categorical, columns= encoder.get_feature_names_out())
X = pd.concat([numericalX, encoded_categorical], axis = 1)

# we need to do train/test split before upsampling, and then only upsample the training set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# for upsampling we need to temporarily concat X_train and y_train
trainset = pd.concat([X_train, y_train], axis=1)


category_1 = trainset[trainset['TARGET_B']==1].sample(len(trainset[trainset['TARGET_B']==0]), replace=True)
print(category_1.shape)
category_0 = trainset[trainset['TARGET_B']== 0 ]
print(category_0.shape)
trainset_new = pd.concat([category_0, category_1], axis = 0)

#randomize the rows
trainset_new = trainset_new.sample(frac =1) 

X_train = trainset_new.drop(['TARGET_B'], axis=1)
y_train = trainset_new['TARGET_B']

print(X_train.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

clf = RandomForestClassifier(max_depth=5, 
                             min_samples_split=20, 
                             min_samples_leaf =20, 
                             max_samples=0.8,    
                             random_state = 42)
clf.fit(X_train, y_train)
print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))

y_pred = clf.predict(X_test)
display(y_test.value_counts())
display(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
clf = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20)
                             

cross_val_scores = cross_val_score(clf, X_train, y_train, cv=10)
print(np.mean(cross_val_scores))
```

```python
# With cross validation the score is slightly better than the test_score above
```

# Comparing different models scores 

```python
from sklearn.tree import DecisionTreeRegressor
model1 = DecisionTreeRegressor()
from sklearn.linear_model import LinearRegression
model2 = LinearRegression()
from sklearn.neighbors import KNeighborsRegressor
model3 = KNeighborsRegressor()
from sklearn.linear_model import LogisticRegression
model4 = LogisticRegression()



model_pipeline = [model1, model2, model3, model4]
model_names = ['Decision Tree Regressor', 'Linear Regression', 'KNN', 'Logreg']
scores = {}
for model, model_name in zip(model_pipeline, model_names):
    mean_score = np.mean(cross_val_score(model, X_train, y_train, cv=5))
    scores[model_name] = mean_score
print(scores)
```
